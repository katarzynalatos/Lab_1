```python
import numpy as np
import matplotlib.pyplot as plt
from LogParser import LogParser
from InputFileValidator import InputFileValidator

file_validator = InputFileValidator("./Logs")
file_validator.validate()
file_list=file_validator.get_list_of_files()
log_parser = LogParser(file_list, "PrChecker.Downs")
date=log_parser.parselog()
name=[]


for x in range(0,14):
    name.append(date[x][0])
file1=[]
file2=[]
file3=[]
file4=[]
for x in range(0,14):
    file1.append(int(date[x][1]))
for x in range(14,28):
    file2.append(int(date[x][1]))
for x in range(28,42):
    file3.append(int(date[x][1]))
for x in range(42,56):
    file4.append(int(date[x][1]))
    

n_groups = 14

#means_men =file1 
std_men = (2, 3, 4, 1, 2)

#means_women = file2
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.3

opacity = 0.6
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, file1, bar_width,
                 alpha=opacity,
                 color='b',
                 #yerr=std_men,
                 error_kw=error_config,
                 label='File 1')

rects2 = plt.bar(index + bar_width, file2, bar_width,
                 alpha=opacity,
                 color='r',
                 #yerr=std_women,
                 error_kw=error_config,
                 label='File 2')
rects3 = plt.bar(index+bar_width, file3, bar_width,
                 alpha=opacity,
                 color='y',
                 #yerr=std_men,
                 error_kw=error_config,
                 label='File 3')

rects4 = plt.bar(index + bar_width, file4, bar_width,
                 alpha=opacity,
                 color='g',
                 #yerr=std_women,
                 error_kw=error_config,
                 label='File 4')

plt.xlabel('Module')
plt.ylabel('Tracks')
plt.title('Tracks amount per module')
plt.xticks(index + bar_width / 4, name, rotation='vertical')
plt.legend()
plt.rcParams["figure.figsize"][0] = 12

plt.tight_layout()
plt.show()
```
